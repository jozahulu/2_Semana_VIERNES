```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=2000  # Numero de Datos aleatorios 
c_r = 25   # Radio del circulo para la variacion de puntos
c_x = 5    # Punto central en el eje x
c_y = 5    # Punto central en el eje y 
angulo = 2 * np.pi * np.random.rand(N)
r = c_r * np.random.rand(N)
```

```python
Ale_anillo = 2*np.pi*np.random.uniform(1.1,1.4,N)
ra_di = np.sqrt(Ale_anillo)              #raiz cuadrada 
x = ra_di * np.cos(angulo) + c_x
y = ra_di * np.sin(angulo) + c_y
plt.plot(x,y,'.',color='green')
plt.xlabel('x')
plt.ylabel('y')
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
# PARA LA PARTE A
dis_datos_menor= []
dis_datos_mayor= []
datos_menor=0
datos_mayor=0
n=0
p_x=np.zeros(50)
p_y=np.zeros(50)
 
```

```python
for i in range(N):
    for j in range(i+1,N):
        if datos_menor   < 50:
            p_x[datos_menor]=x[j]
            p_y[datos_menor]=y[j]
            d_d =1* distancia(x[i],x[j],y[i],y[j])
            dis_datos_menor.append(d_d)
            datos_menor =datos_menor +1
            datos_menor =datos_menor 
            #print('px',p_x)
            #print('py',p_y) 
        else:
            d_d = 2*distancia(x[i],x[j],y[i],y[j])
            #d_a = distancia(p_x[i],p_x[j],p_y[i],p_y[j])
            dis_datos_mayor.append(d_d)
            #dis_aleatorios.append(d_a)
            datos_mayor =datos_mayor +1
            datos_mayor =datos_mayor
            
            
print('datos_menor',datos_menor)
print('datos_mayor',datos_mayor)     
print('px',p_x)
print('py',p_y)  
```

```python
data_n, data_bin,m = plt.hist(datos_menor,bins=20,color='red')
```

```python
ran_n, ran_bin, p = plt.hist(dis_datos_mayor,bins=20,color='green')
```

```python
plt.plot(data_bin[1:],(data_n/ran_n) -1, label=('$\dfrac{DD(r)}{RR(r)}-1$'),color='red')
plt.legend()
```

```python
dit_da = []
```

```python
for i in range(50):
    for j in range(N-50):
        d_dr = distancia(x[j+50],p_x[i],y[j+50],p_y[i])
        dit_da.append(d_dr)
```

```python
da_n, da_bin, q = plt.hist(dit_da,bins=20,color='blue')
```

```python
plt.plot(data_bin[1:],(data_n*ran_n)/(da_n)**2,label=('$\dfrac{DD(r)RR(r)}{DR(r)^2}$'))
plt.legend()
```

```python
plt.title('Landy-Szalay')
plt.plot(data_bin[1:],1.+ data_n/ran_n -2*da_n/ran_n, label=('Landy-Szalay=$1+\dfrac{DD(r)}{RR(r)}-2\dfrac{DR(r)}{RR(r)}$'))
```

```python
plt.plot(data_bin[1:],((data_n)/(da_n))-1,label=('Davis-Peebles=$\dfrac{DD(r)}{DR(r)}-1$'))

plt.legend()
```
