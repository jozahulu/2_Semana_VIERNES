```python
import numpy as np 
from scipy.fftpack import fftfreq
from scipy.fftpack import ifftn
import matplotlib.pyplot as plt
import matplotlib.cm as cm


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=2000  # Numero de Datos aleatorios 
c_r = 25   # Radio del circulo para la variacion de puntos
c_x = 5    # Punto central en el eje x
c_y = 5    # Punto central en el eje y 
angulo = 2 * np.pi * np.random.rand(N)
r = c_r * np.random.rand(N)
```

```python
Ale_anillo = 2*np.pi*np.random.uniform(1.1,1.4,N)
ra_di = np.sqrt(Ale_anillo)              #raiz cuadrada 
x = ra_di * np.cos(angulo) + c_x
y = ra_di * np.sin(angulo) + c_y
print(len(x))
print(x[0])
plt.plot(x,y,'.',color='green')
plt.xlabel('x')
plt.ylabel('y')
```

```python
xf=fft(x)  ###Transformada de Fourier
yf=fft(y)
plt.subplot(1,3,1)
plt.title('Grafica del conjunto')
plt.plot(np.abs(xf),np.abs(yf),'.',color='green')
plt.xlabel('x')
plt.ylabel('y')
ax=plt.gca()
ax.set_ylim([0,180])
ax.set_xlim([0,180])

plt.subplot(1,3,3)
plt.title('Grafica total')
plt.plot(np.abs(xf),np.abs(yf),'.',color='green')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# PARA DIFERENTES ORDENAMIENTOS NO FUNCIONA PERO ES UN Camino
```

```python
N = 30 # Solo para 30 valores
f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, sharex='col', sharey='row') # Sirve para ordenar dibujos
xf = np.zeros((N,N))  # Representacion vectorial
xf[0, 5] = 1
xf[0, N-5] = 1
Z = ifftn(xf)  # Transformada de fourier
ax1.imshow(xf, cmap=cm.Reds)  # Determinacion de ordenamiento
ax4.imshow(np.real(Z), cmap=cm.gray)
xf = np.zeros((N, N))
xf[5, 0] = 1
xf[N-5, 0] = 1
Z = ifftn(xf)   # Transformada inversa
ax2.imshow(xf, cmap=cm.Reds)
ax5.imshow(np.real(Z), cmap=cm.gray)
xf = np.zeros((N, N))
xf[5, 10] = 1
xf[N-5, N-10] = 1
Z = ifftn(xf)
ax3.imshow(xf, cmap=cm.Reds)
ax6.imshow(np.real(Z), cmap=cm.gray)
plt.title('los ordenamientos no funcionan para mas valores MAYORES')
plt.show()
```
